## Distribución Uniforme

Para generar números aleatorios uniformes entre dos valores dados usando scipy, podemos usar el siguiente código o el código que hemos
discutido en clase:

```python
from scipy.stats import uniform

# Definir los límites inferior y superior de la distribución uniforme
limite_inferior = 10
limite_superior = 20

# Crear una instancia de la distribución uniforme entre los límites dados
distribucion_uniforme = uniform(loc=limite_inferior, scale=limite_superior-limite_inferior)

# Generar números aleatorios uniformemente distribuidos
num_aleatorios = distribucion_uniforme.rvs(size=1000)
```

**Tarea:**

1. Basándote en el código anterior, escribe una función que genere números aleatorios uniformemente distribuidos. Hay varios requisitos para tu función:
    * Debe aceptar 3 parámetros: 
        * `bottom` - el límite inferior de los números generados
        * `ceiling` - el límite superior de los números generados
        * `count` - cuántos números generar
    * Debe devolver una matriz de números aleatorios distribuidos uniformemente

2. Llama a tu función con 2 conjuntos de parámetros a continuación:
    * bottom=10, ceiling=15, count=100
    * bottom=10, ceiling=60, count=1,000

3. Traza las distribuciones uniformes generadas anteriormente utilizando histogramas, donde el eje x es el valor y el eje y es el recuento. Que el número de intervalos del histograma sea 10.

El resultado debe ser como el siguiente:



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

def generar_uniforme(bottom, ceiling, count):
    # Crear una instancia de la distribución uniforme entre los límites dados
    distribucion_uniforme = uniform(loc=bottom, scale=ceiling-bottom)
    
    # Generar números aleatorios uniformemente distribuidos
    num_aleatorios = distribucion_uniforme.rvs(size=count)
    
    return num_aleatorios

# Llamar a la función con los primeros parámetros: bottom=10, ceiling=15, count=100
datos1 = generar_uniforme(bottom=10, ceiling=15, count=100)

# Llamar a la función con los segundos parámetros: bottom=10, ceiling=60, count=1000
datos2 = generar_uniforme(bottom=10, ceiling=60, count=1000)

# Crear histogramas
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.hist(datos1, bins=10, edgecolor='k', alpha=0.7)
plt.title('Distribución Uniforme\n(bottom=10, ceiling=15, count=100)')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')

plt.subplot(1, 2, 2)
plt.hist(datos2, bins=10, edgecolor='k', alpha=0.7)
plt.title('Distribución Uniforme\n(bottom=10, ceiling=60, count=1000)')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()

```

How are the two distributions different?

## Distribución Exponencial

1. Utilizando `numpy.random.exponential`, crea una función que devuelva una lista de números distribuidos exponencialmente con la media de 10. 

1. Utiliza la función para generar dos secuencias de números de tamaño 10 y 100.

1. 2. Traza las distribuciones como histogramas con el número de bins como 100.

El resultado debe ser como el siguiente:



```python
import numpy as np
import matplotlib.pyplot as plt

# Crear la función que genera números distribuidos exponencialmente
def generar_numeros_exponenciales(media, tamaño):
    return np.random.exponential(scale=media, size=tamaño)

# Generar dos secuencias de números con tamaños 10 y 100
media = 10
secuencia_10 = generar_numeros_exponenciales(media, 10)
secuencia_100 = generar_numeros_exponenciales(media, 100)

# Configurar la figura para los histogramas
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Trazar el histograma para la secuencia de tamaño 10
ax1.hist(secuencia_10, bins=100, color='blue', edgecolor='black')
ax1.set_title('Histograma de Secuencia de Tamaño 10')
ax1.set_xlabel('Valor')
ax1.set_ylabel('Frecuencia')

# Trazar el histograma para la secuencia de tamaño 100
ax2.hist(secuencia_100, bins=100, color='green', edgecolor='black')
ax2.set_title('Histograma de Secuencia de Tamaño 100')
ax2.set_xlabel('Valor')
ax2.set_ylabel('Frecuencia')

# Mostrar los histogramas
plt.tight_layout()
plt.show()
```

¿En qué se diferencian ambas distribuciones?

La media cambia, por lo que la distribución también cambia. 

## Distribución Exponencial de datos reales

Supongamos que el tiempo que uno pasa en un banco se distribuye exponencialmente con media de 10 minutos (es decir, λ = 1/10). ¿Cuál es la probabilidad de que un cliente pase menos de quince minutos en el banco? 

Escribe un código en python para resolver este problema

```python
import math

# Parámetro lambda
lambd = 1 / 10

# Tiempo en minutos
x = 15

# Calculamos la probabilidad P(X < 15)
probabilidad = 1 - math.exp(-lambd * x)

print(f"La probabilidad de que un cliente pase menos de 15 minutos en el banco es: {probabilidad:.4f}")
```
